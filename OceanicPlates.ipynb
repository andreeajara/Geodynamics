```python
%pylab inline
import math
import numpy as np
import scipy.special as sp
from matplotlib import pyplot
```

```python
k=10**-6 #m^2/s
def ridgeT(x,y,v): #v in cm/yr, y and x in km
    T=sp.erfc(abs(y*1000)/(2*sqrt(k*abs(x*1000)*100*365*24*3600/v)))*-1300 + 1573.15 #K
    return T
```

```python
x=linspace(-150,150,500)
y=linspace(0,-80,1000)
X,Y=np.meshgrid(x,y)
```

```python
Z1 = ridgeT(X,Y,0.5)
Z2 = ridgeT(X,Y,4)
Z3 = ridgeT(X,Y,10)

fig, axes = plt.subplots(3,1,figsize=(27,7.5))
plt.tight_layout(4)

subplot(3, 1, 1)
im = plt.imshow(Z1,extent=[x.min(), x.max(), y.min(), y.max()])
plt.title('Plate velocity: 0.5 cm/yr',fontsize=12)
plt.ylabel('z (km)', fontsize=10)
plt.xlabel('x (km)', fontsize=10)

subplot(3, 1, 2)
plt.imshow(Z2,extent=[x.min(), x.max(), y.min(), y.max()])
plt.title('Plate velocity: 4 cm/yr',fontsize=12)
plt.ylabel('z (km)', fontsize=10)
plt.xlabel('x (km)', fontsize=10)

subplot(3, 1, 3)
plt.imshow(Z3,extent=[x.min(), x.max(), y.min(), y.max()])
plt.title('Plate velocity: 10 cm/yr',fontsize=12)
plt.xlabel('x (km)', fontsize=10)
plt.ylabel('z (km)', fontsize=10)

cbaxes = fig.add_axes([0.63, 0.1, 0.01, 0.81]) 
cb = plt.colorbar(im, cax = cbaxes)
cb.set_label('Temperature (K)', fontsize=12, rotation=270, labelpad=19)
```

```python
def yBL(x,v): #x in km, v in cm/yr
    y=-2.32*sqrt(k*abs(x*1000)*(100*365*24*3600)/v)
    return y/1000 #in km
```

```python
figure(figsize=(12,4))
plot(x,yBL(x,0.5),c='r',label=r'0.5 cm/yr')
plot(x,yBL(x,4),c='g',label=r'4 cm/yr')
plot(x,yBL(x,10),c='b',label=r'10 cm/yr')
title("Thickness of the thermal lithosphere",fontsize=15)
xlabel('x (km)', fontsize=12)
ylabel('z (km)', fontsize=12)
legend(fontsize=12, loc='lower center')
ylim(-80,0)
```

```python
rhom=3300 #kg/m^3
rhow=1000 #kg/m^3
alpha=1.6*10**-5 #/°C
deltaT=1300
def water(x,v): #x in km, v in cm/yr
    w=(-rhom/(rhom-rhow))*alpha*deltaT*2*sqrt(k*abs(x*1000)*(100*365*24*3600)/v)/sqrt(pi)
    return w/1000 #in km
```

```python
plt.subplots(3,1,figsize=(10,12))
plt.tight_layout(4)

subplot(3, 1, 1)
plot(x,water(x,0.5), label=r'Bathymetry')
plot(x,yBL(x,0.5), label=r'Plate thickness')
title("Cross section of oceanic lithosphere. Half-spreading rate: 0.5 cm/yr",fontsize=15)
ylim(-80,0)
xlabel('x (km)', fontsize=12)
ylabel('z (km)', fontsize=12)
legend(fontsize=10, loc='lower right')

subplot(3, 1, 2)
plot(x,water(x,4), label=r'Bathymetry')
plot(x,yBL(x,4), label=r'Plate thickness')
title("Cross section of oceanic lithosphere. Half-spreading rate: 4 cm/yr",fontsize=15)
ylim(-80,0)
xlabel('x (km)', fontsize=12)
ylabel('z (km)', fontsize=12)
legend(fontsize=10, loc='lower right')

subplot(3, 1, 3)
plot(x,water(x,10), label=r'Bathymetry')
plot(x,yBL(x,10), label=r'Plate thickness')
title("Cross section of oceanic lithosphere. Half-spreading rate: 10 cm/yr",fontsize=15)
ylim(-80,0)
xlabel('x (km)', fontsize=12)
ylabel('z (km)', fontsize=12)
legend(fontsize=10, loc='lower right')
```

```python
R=8.314 #J / K mol
ea=523*10**3 #J mol
C1=4.2*10**5 #MPa^-3 s^-1
E=7*10**4 #Mpa
def tau(T,sigma): #T in K, sigma in MPa
    t=((3*exp(ea/(R*T)))/(2*E*C1*sigma**2)) #Turcotte & Schubert, 2007, p. 612.
    return t/(365*24*3600) #in yr
```

```python
T=linspace(300,1600,500)
figure(figsize=(12,6))
plot(T,tau(T,10), label=r'$\sigma = 10\ MPa $')
plot(T,tau(T,100), label=r'$\sigma = 10^2 MPa $')
plot(T,tau(T,1000), label=r'$\sigma = 10^3 MPa $')
title("Relaxation time as a function of temperature",fontsize=17)
xlabel(r'$T (K)$', fontsize=15)
ylabel(r'$ \tau_r (yr) $', fontsize=15)
pyplot.yscale('log')
legend(fontsize=15, loc='lower left')
xlim(300,1600)
ylim(0,10**9)
```
